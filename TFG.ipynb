```python
import pandas as pd

# Cargar el CSV
file_path = '/content/drive/MyDrive/TFG/Obfuscated-MalMem2022.csv'
df = pd.read_csv(file_path)

# Ver las primeras filas
df.head()
```

```python
df.shape
```

```python
# Quitar puntos de miles y convertir a float
def limpiar_columna(col):
    try:
        return pd.to_numeric(col.str.replace('.', '', regex=False), errors='coerce')
    except:
        return col  # Si falla (por no ser str), se devuelve tal cual

# Aplicamos a todas las columnas numéricas que parezcan estar mal formateadas
for col in df.columns:
    if df[col].dtype == 'object':
        df[col] = limpiar_columna(df[col])

# Mostramos tipos de datos tras limpieza
df.dtypes
```

```python
# Comprobamos cuántos valores nulos hay por columna
print(df.isnull().sum())

# También el porcentaje de valores nulos por columna (opcional)
print((df.isnull().sum() / len(df)) * 100)
```

```python
# Veamos qué clases hay y cuántas veces aparece cada una
print(df['Category'].value_counts())
```
